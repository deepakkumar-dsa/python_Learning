#### create a function to calculate the sum of given of values using with input and with return after returning divide that value by 5


```python
def sum_div(n1,n2):
    s = 0
    for i in range(n1, n2+1,1):
       s = s + 1
    return s

```

```python
def fact(n):
    if n==1:
        return 1

    else:
        return n*fact(n-1)
    
```

```python
def fib(n):
    if n<=1:
        return n 
    return fib(n-1)+fib(n-2)
fib(6)
for i in range(6):
    print(fib(i))
```
