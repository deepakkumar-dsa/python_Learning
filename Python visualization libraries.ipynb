 ##  Visualiztion

#### why data visualization?
* Easy understanding of data it save more time because we can quickly analyze the data

#### Python based visualization libraries
* Matplotlib - more customized
* seaborn
* plot

#### Name of column and rows
* Columns : features
* rows : records/observations

#### Univariate analysis
* To understand the quality of the one feature counplot,histogram, box plot, scatterplot.

#### bivariate analysis 
* To understand the relationship between two features barplot,scatterplot.

#### Multivariate analysis 
* To understand the relationship between more than 2 features heatmap,pairplot.

```python
from matplotlib import pyplot as plt
```

```python
plt.figure(figsize=(5,5))
plt.hist(x='mpg',data=mt_cars_data)
plt.title("MPG Distribution")
plt.xlabel("MPG")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(x='wt',data=mt_cars_data)
plt.title("wt Distribution")
plt.xlabel("wt")
plt.ylabel("Frequency")
plt.show()
```

```python
for i in mt_cars_data.columns:
    if mt_cars_data.dtypes[i]=="float64":
        plt.figure(figsize=(8,5))
        plt.hist(x=i,data=mt_cars_data)
        plt.title("Distribution")
        plt.xlabel(i)
        plt.ylabel("Frequency")
        plt.show()
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='cyl',y=None,data=mt_cars_data,hue='cyl')
plt.title("cyl Distribution")
plt.show()
```

```python
import seaborn as sns
```

```python
sns.barplot(x='gear',y='mpg',data=mt_cars_data)
plt.title("gear vs mpg relationship")
plt.xticks(rotation=20)
plt.show()
```

```python
sns.barplot(x='wt',y='mpg',data=mt_cars_data)
plt.title("wt vs mpg relationship")
plt.xticks(rotation=90)
plt.show()
```

```python
sns.scatterplot(x='wt',y='mpg',data=mt_cars_data)
plt.xlabel("wt")
plt.ylabel("mpg")
plt.title('wt vs mpg relationship')
plt.xticks(rotation=20)
plt.show()
```

```python
# multivariate analysis
sns.pairplot(data=mt_cars_data)
plt.show()
```

```python
#correlation matrix
corr=mt_cars_data.corr().round(2)
corr
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(data= corr,annot= True)
plt.show()
```

```python
#import matplotlib.pyplot as plt
plt.pie(x=mt_cars_data['cyl'].value_counts(),labels=[8,4,6],autopct='%0.2f%%',explode=[0,0.03,0])
plt.show()
```

```python
#Boxplot - Data Distribution and detects the outliers
plt.boxplot(x='wt',data= mt_cars_data,vert=False)
plt.show()
```

```python
x=[-3,5,7]
y=[10,2,5]
plt.figure(figsize=(10,5),facecolor="black")
plt.plot(x,y,color="black",linewidth=5,linestyle='-.')
plt.scatter(x,y,color="red",s=100,marker='*')
plt.xlabel("X Axis" ,size=20,color="red")
plt.ylabel("Y Axis" ,size=20,color="red")
plt.title("My First plot in python", size=20,color="green")
plt.xticks(range(-4,10),color="red")
plt.yticks(range(1,11),color="red")
plt.text(-3.5,9.5, "outlier",size=12,color="red")
plt.show()
```
